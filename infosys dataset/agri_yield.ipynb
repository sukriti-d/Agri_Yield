```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('crop_yield.csv')
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
# Crop Yield distribution
plt.figure(figsize=(8,5))
df['Yield_tons_per_hectare'].hist(bins=50)
plt.xlabel('Crop Yield (tons/hectare)')
plt.ylabel('Frequency')
plt.title('Crop Yield Distribution')
plt.show()
```

```python
# Crop Type 
plt.figure(figsize=(8,5))
df['Crop'].value_counts().plot(kind='bar')
plt.xlabel('Crop Type')
plt.ylabel('Count')
plt.title('Count of Each Crop Type')
plt.show()
```

```python
# mean crop yield by region
region_yield = df.groupby('Region')['Yield_tons_per_hectare'].mean()
plt.figure(figsize=(8,5))
region_yield.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Mean Crop Yield (tons/hectare)')
plt.title('Mean Crop Yield by Region')
plt.show()
```
